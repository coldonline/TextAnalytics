```python
import GetOldTweets3 as got
import numpy as np
import pandas as pd
```

```python
#Choose number of cities --> in order of population
n = 100
#Choose number of tweets per city
max_tweets = 500
```

```python
#import city data ordered by population 
cities_raw = pd.read_csv("us_cities.txt", delimiter = ",")
cities = cities_raw["city"]
cities_mini = pd.Series.tolist(cities[:n])
cities_mini
```

```python
#use get old tweets to search tweets based on number of cities selected
tweetCriteria_list = []
for city in cities_mini:
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('#NFL')\
                                           .setSince("2018-01-01")\
                                           .setUntil("2019-11-23")\
                                            .setNear(city)\
                                           .setMaxTweets(max_tweets)
    tweetCriteria_list.append(tweetCriteria)
```

```python
#create dictionary of twitter info by city
tweet_dict = {}
counter = 0
for criteria, city in zip(tweetCriteria_list, cities_mini):
    tweets = got.manager.TweetManager.getTweets(criteria)
    tweet_dict[city] = tweets
    counter += 1
    print("Completed city: ", counter)
```

```python
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in tweet_dict.items() ]))
```

```python
df['tweet_count'] = df.index
df.head()
```

```python
df = pd.melt(df, id_vars=["tweet_count"], var_name='City', value_name='got_criteria')
df = df.dropna()
df.head()
```

```python
#Create columns with tweet information
df["tweet"] = df["got_criteria"].apply(lambda x: x.text)
df["date"] = df["got_criteria"].apply(lambda x: x.date)
df["hashtags"] = df["got_criteria"].apply(lambda x: x.hashtags)
df["link"] = df["got_criteria"].apply(lambda x: x.permalink)
```
